```python
# Preprocess the data
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize

target = []
images = [] # 2D
flat_data = []
DATADIR = "/content/images"
CATEGORIES = ['the taj mahal','lotus temple','eiffel tower','colosseum']

for category in CATEGORIES:
  #print(category)
  class_num = CATEGORIES.index(category)
  #print(class_num)
  path = os.path.join(DATADIR,category)
  #print(path)
  for img in os.listdir(path):
    imga = imread(os.path.join(path,img))
    #print(imga.shape)
    #plt.imshow(imga)
    img_resize = resize(imga,(150,150,3))
    flat_data.append(img_resize.flatten())
    images.append(img_resize)
    target.append(class_num)
flat_data = np.array(flat_data)
target = np.array(target)
images = np.array(images)
```

```python
import numpy as np
unique,count = np.unique(target,return_counts=True)
plt.bar(CATEGORIES,count)
```

```python
#Transfer Learning
import tensorflow as tf
from keras import layers
from keras import Model
base_model = tf.keras.applications.InceptionV3(weights='imagenet', include_top = False)
```

```python
from tensorflow import keras
from keras.applications.inception_v3 import InceptionV3
```

```python
model = InceptionV3(weights='imagenet', include_top = False)
model.summary()
```

```python
# Train the model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flat_data, target, test_size = 0.3, random_state = 100)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = [
              {'C':[1,10,100,1000],'kernel':['linear']},
              {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']},
]

svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
model.compile(loss = 'sparse_categorical_crossentropy',optimizer = 'Adam',metrics = ['accuracy'])
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
%%writefile app.py
import streamlit as st
import numpy as np
from skimage.io import imread
from skimage.transform import resize
import pickle 
from PIL import Image
st.set_option('deprecation.showfileUploaderEncoding', False)
st.title('Image Classifier using Machine Learning')
st.text('Upload the Image')

model = pickle.load(open('img_model.p','rb'))

uploaded_file = st.file_uploader("Choose an image...")
if uploaded_file is not None:
  img = Image.open(uploaded_file)
  st.image(img,caption='Uploaded Image')

  if st.button('PREDICT'):
    CATEGORIES = ['colosseum','eiffel tower','lotus temple','the taj mahal']
    st.write('Result...')
    flat_data=[]
    img = np.array(img)
    img_resized = resize(img,(150,150,3))
    flat_data.append(img_resized.flatten())
    flat_data = np.array(flat_data)
    y_out = model.predict(flat_data)
    y_out = CATEGORIES[y_out[0]]
    st.title(f' PREDICTED OUTPUT: {y_out}')
    q = model.predict_proba(flat_data)
    for index, item in enumerate(CATEGORIES):
      st.write(f'{item} : {q[0][index]*100}%')
```
